### Data Preprocessing

```python
stop_words = stopwords.words('english')
```

```python
def clean_text(text):
    text = unidecode(text)
    text = str(text).lower()
    text = str(text).translate(str.maketrans(string.punctuation,' '*len(string.punctuation)))
    text = str(text).translate(str.maketrans('','', string.digits))    
    text = text.split()
    text = " ".join([stem(i) for i in text]) # if len(i)>2 and len(i)<15]  #  if i not in stop_words
    return text
```

```python
# Dropping Unnecessary columns, and rows
df = df.drop(['post_id', 'created_date', 'if_1', 'if_2'], axis=1)
df = df.drop(df[df['comment'].isnull()].index).reset_index(drop=True)
```

```python
(df=='none').sum()
```

```python
# Replacing 'none' with nan
df = df.replace('none','')
df[['race', 'religion', 'gender']] = df[['race', 'religion', 'gender']].replace(np.nan,'')
```

```python
df.isna().sum()
```

```python
# Adding column values to the comment
df['comment'] = df['comment'] + df[['race', 'religion', 'gender']].apply(lambda x: ' '.join(x), axis=1)
```

```python
# Adding the word 'disability' to the comment where column disability has value True, as this will be incorporated as a feature in vectorization Step
df.loc[df['disability'],'comment'] = df.loc[df['disability'],'comment'] + ' disability'
```

```python
df = df.drop(['race', 'religion', 'gender', 'emoticon_1', 'emoticon_2', 'emoticon_3', 'disability'], axis=1)
```

```python
train_df, val_df = train_test_split(df, test_size=.2, stratify=df['label'], random_state=42)
```

```python
train_df = train_df.groupby('label', as_index=False).agg(upvotes=('upvote','sum'), downvotes=('downvote','sum'), comment_text=('comment', ' '.join))
```

```python
cv = CountVectorizer(preprocessor=clean_text, ngram_range=(1,2), max_features=50000, min_df=2)
```

```python
matrix = cv.fit_transform(train_df['comment_text'])
```

```python
val_m = cv.transform(val_df['comment'])
```

```python
from collections import Counter
```

```python
doc_freq = np.array([i[1] for i in sorted(Counter(matrix.nonzero()[1]).items())])
```

```python
doc_len = np.array([91338, 49952, 12734,  4375])
```

```python
avgdl = round(np.mean(doc_len),0)
```

```python
# https://en.wikipedia.org/wiki/Okapi_BM25
idf = np.log((4 - doc_freq + 0.5))-np.log(doc_freq+0.5)
```

```python
for i in range(len(idf)):
    if idf[i]<0:
        idf[i]=0.25
```

```python
b = 0.22
k = 240
m2 = csr_matrix(idf *matrix.toarray()*(k+1) / (matrix.toarray() + k * (1 - b + b * doc_len.reshape(-1,1) / avgdl)))

```

```python
y_hat = np.array(np.argmax(val_m@m2.T,axis=1)).flatten()
```

```python
confusion_matrix(val_df['label'].values, y_hat, labels=[0,1,2,3])
```

```python
accuracy_score(val_df['label'].values, y_hat)
```
